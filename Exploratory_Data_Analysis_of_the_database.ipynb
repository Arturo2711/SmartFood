Since the main attributes are:

*   kcal
*   protein

I'm going to remove all attributes except those, since the goal of the project is to implement a recipe generator. It makes sense




```python
mainDf = df[['nombre_del_alimento', 'group', 'energ_kcal', 'carbohydrt', 'lipid_tot', 'protein']]
```

```python
mainDf.describe()
```

## Look for correlations

```python
corr_matrix = mainDf[['group', 'energ_kcal', 'carbohydrt', 'lipid_tot', 'protein']].corr()
```

```python
print(corr_matrix)
```

Let's plot it

```python
from pandas.plotting import scatter_matrix
```

```python
attributes = ['group', 'energ_kcal', 'carbohydrt', 'lipid_tot', 'protein']
scatter_matrix(mainDf[attributes], figsize=(30, 50))
plt.show()
```

## Clean the data

Set the missing values to some value

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='constant', fill_value=0)
```

Since the imputer only work with numerical attributes

```python
mainDf_numerical = mainDf[attributes]
```

```python
imputer.fit(mainDf_numerical)
```

```python
imputer.statistics_
```

```python
final_Data_Set = imputer.transform(mainDf_numerical)
```

```python
df_inputed = pd.DataFrame(final_Data_Set)
```

```python
df_inputed.info()
```

```python
final_df = df_inputed.reset_index(drop=True)
final_df['name'] = mainDf['nombre_del_alimento']
final_df.to_csv('/content/drive/MyDrive/SmartFood/data_base.csv')
```
